## Project

Classes

```
class Question:
    def __init__(self,prompt,ans):
        self.prompt = prompt
        self.ans = ans
```

Papers

```
phy = ['1.The phase difference between the alternating current and emf is TE/2. Which ofthe following cannot be the constituent of the circuit?\n(a) C alone\n(b) L alone\n(c) L and C\n(d) R and L\n\n',
           '2.In an LCR-series ac circuit, the voltage across each of the component L, C and R is 50 V. The voltage across the LC-combination will be\n(a) 50 V\n(b) 50√2 V\n(c) 100 V\n(d) zero\n\n',
          '3. An ac circuit has a resistance of 12 ohm and an impedance of 15 ohm. The power factor of the circuit will be\n(a) 0.8\n(b) 0.4\n(c) 0.125\n(d) 1.25\n\n',
           '4.The core of any transformer is laminated so as to\n(a) reduce the energy loss due to eddy currents.\n(b) make it light weight.\n(c) make it robust and strong.\n(d) increase the secondary voltage.\n\n',
'5. If coil is open, then L and R becomes\n(a) infinity, zero\n(b) zero, infinity\n(c) infinity, infinity.\n(d) zero, zero\n\n'
]

math = ["1.Given set A ={1, 2, 3} and a relation R = {(1, 2), (2, 1)}, the relation R will be\n(a)reflexive if (1, 1) is added\n(b)symmetric if (2, 3) is added\n(c)transitive if (1, 1) is added\n(d)symmetric if (3, 2) is added\n\n",
        "2. What is value of i ?\n (a) -1\n (b) -1^0.5\n (c)0\n\n",
        "3. What is value of 0/0?\n (a) Infinity\n (b) 0\n (c)Undefined\n\n",
'4. Let A be a square matrix of order 2 × 2,then |KA| is equal to\n(a) K|A|\n(b) K²|A|\n(c) K3|A|\n(d) 2K|A|\n\n',
 '5. Determinant is a number associated to a matrix.\n (a) True\n (b)False\n\n'
 ]

chem = ['1. The formation of micelles takes place only above\n(a) Inversion temperature\n(b) Boyle’s temperature\n(c) Critical temperature\n(d) Kraft temperature\n\n',
'2.Colloidion is 4% solution of which one of the following in alcohol-ether mixture.\n(a) Nitroglycerin\n(b) Cellulose acetate\n(c) Glycol dinitrate\n(d) Nitrocellulose\n\n',
'3. A plot of log xm vs log p for adsorption of gas on a solid gives in straight line with slope equal to\n(a)\n(b) 1n\n(c) log k\n(d) -log k\n\n',
'4.The protective power of lyophilic colloidal sol is expressed in terms of\n(a) coagulation value\n(b) gold number\nc) CMC\n(d) oxidation numbers\n\n',
'5. Formula of water is\n (a)H20\n (b) X4O\n (c)H30 \n (d) H2SO4\n\n '
]

phyques = [Question(phy[0], 'c'), Question(phy[1],'b'),Question(phy[2],'b'),Question(phy[3],'a'),Question(phy[4],'a')]
mathques = [Question(math[0], 'b'), Question(math[1],'a'),Question(math[2],'c'),Question(math[3],'b'),Question(math[4],'a')]
chemques = [Question(chem[0], 'd'), Question(chem[1],'d'),Question(chem[2],'b'),Question(chem[3],'b'),Question(chem[4],'b')]
```

Quiz

```
def run(physques,mathques,chemques):
    print("Welcome student please give the Physics quiz\n\n")
    score1 = 0
    for x in physques:
        ans1 = input(x.prompt)
        if ans1 == x.ans:
            score1 += 10
    clear_output()
    print("Now its time for maths quiz\n\n")
    score2 = 0
    for y in mathques:
        answer = input(y.prompt)
        if answer == y.ans:
            score2+=10
    clear_output()
    print("Finally give the chems quiz\n\n")
    score3 = 0
    for z in chemques:
        ans3 = input(z.prompt)
        if ans3 == z.ans:
            score3+=10
    clear_output()
    return score1,score2,score3
    
```

Main Function

```
marks = []
n = int(input("How many students are giving test ?"))
clear_output()
for x in range(n):
    m1 = []
    name = input("Please write your name")
    gen = input("Please mention your gender(M/F)")
    course = input("Did you took an online course from platforms like Edx/Coursera(Y/N)")
    s1,s2,s3 = run(phyques,mathques,chemques)
    m1.extend((name,gen,course,s1,s2,s3))
    marks.append(m1)
```

Conversion into dataframe

```
df = pd.DataFrame(data = marks, columns = ['Name','Gender','Course','Phy','Maths','Chem'])
df.set_index('Name',inplace = True)
df['Final Percentage']  = ((df['Maths'] + df['Phy'] + df['Chem'] )/150)*100
df
```
